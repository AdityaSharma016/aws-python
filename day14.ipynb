```python
import matplotlib.pyplot as plt

# -------------------- Data --------------------
fruits = ["Apple", "Banana", "Carrot", "Mango"]
values = [30, 50, 70, 5]

experience = [1, 2, 3, 4, 5]
salary = [100, 200, 300, 400, 500]
bonus = [20, 40, 60, 80, 100]

# -------------------- Create 2x2 Subplots --------------------
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# -------------------- Pie Chart --------------------
ax[0, 0].pie(values, labels=fruits, autopct='%1.1f%%')
ax[0, 0].set_title("Pie Chart")

# -------------------- Bar Chart --------------------
ax[0, 1].bar(fruits, values, color='skyblue')
ax[0, 1].set_title("Bar Chart")
ax[0, 1].set_xlabel("Fruits")
ax[0, 1].set_ylabel("Values")

# -------------------- Line Chart --------------------
ax[1, 0].plot(experience, salary,
              marker='o',
              color='green',
              linewidth=2)
ax[1, 0].set_title("Line Chart")
ax[1, 0].set_xlabel("Experience")
ax[1, 0].set_ylabel("Salary")

# -------------------- Multiple Line Chart --------------------
ax[1, 1].plot(experience, salary,
              marker='o',
              label='Salary')

ax[1, 1].plot(experience, bonus,
              marker='s',
              label='Bonus')

ax[1, 1].set_title("Multiple Line Chart")
ax[1, 1].set_xlabel("Experience")
ax[1, 1].set_ylabel("Amount")
ax[1, 1].legend()

# -------------------- Adjust Layout --------------------
plt.tight_layout()
plt.savefig("adii")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# -------------------- Data --------------------
fruits = ["Apple", "Banana", "Carrot", "Mango"]
values = [30, 50, 70, 5]

experience = [1, 2, 3, 4, 5]
salary = [100, 200, 300, 400, 500]
bonus = [20, 40, 60, 80, 100]

# -------------------- Create 2x2 Subplots --------------------
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# -------------------- Pie Chart --------------------
ax[0, 0].pie(values, labels=fruits, autopct="%1.1f%%")
ax[0, 0].set_title("Pie Chart")

# -------------------- Bar Chart --------------------
ax[0, 1].bar(fruits, values, color="skyblue")
ax[0, 1].set_title("Bar Chart")
ax[0, 1].set_xlabel("Fruits")
ax[0, 1].set_ylabel("Values")

# -------------------- Line Chart --------------------
ax[1, 0].plot(experience, salary,
              marker="o",
              color="green",
              linewidth=2)
ax[1, 0].set_title("Line Chart")
ax[1, 0].set_xlabel("Experience")
ax[1, 0].set_ylabel("Salary")

# -------------------- Multiple Bar Chart --------------------
x = np.arange(len(experience))   # [0,1,2,3,4]
width = 0.35

ax[1, 1].bar(x - width/2, salary, width,
             label="Salary", color="skyblue")

ax[1, 1].bar(x + width/2, bonus, width,
             label="Bonus", color="salmon")

ax[1, 1].set_title("Multiple Bar Chart")
ax[1, 1].set_xlabel("Experience")
ax[1, 1].set_ylabel("Amount")
ax[1, 1].set_xticks(x)
ax[1, 1].set_xticklabels(experience)
ax[1, 1].legend()

# -------------------- Adjust Layout --------------------
plt.tight_layout()
plt.show()
```
