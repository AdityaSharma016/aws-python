```python
#create Data Frame
data ={
    "name": ["aditya","yadvav","yuoo","nancy"],
    "age": [10,20,30,40]
}
df=pd.Series(data)
print(df)
df = pd.DataFrame(data)
print(df)

```

```python
data=[['nikhil',10],["aditya",50]]
df=pd.DataFrame(data,columns=['name','age'])
print(df)

#add indexing name

df=pd.DataFrame(data,index=['a','b'])
print(df)
```
