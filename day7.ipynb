```python
class User1:
    def __init__(self, pen, pencil):
        self.pen = pen
        self.pencil = pencil

    def pen_get(self):
        print("pen is working", self.pen)

    def pencil_get(self):
        print("pencil is working", self.pencil)


class User2(User1):
    def __init__(self, pen, pencil, eraser):
        super().__init__(pen, pencil)
        self.eraser = eraser


u = User2("blue", "black", "doms")
u.pen_get()
```

```python
class Aditya:
    def __init__(self,bike):
        self.bike=bike
    def usebike(self):
        print("bike used by: ",self.bike)
class Sharma(Aditya):
    def __init__(self, name, bike):
        #self.bike=bike
        self.name=name
        super().__init__(bike)#super is used to get the properrtitis of the parent class 
    def usename(self):
        print("name is: ",self.name)

u = Sharma("Aditya Sharma", "honda")
u.usename()
u.usebike()
        
```

```python
class Math:
    def pow(self, a, b):
        print(a ** b)

    def factorial(self, a=6):
        count = 1
        for i in range(1, a + 1):
            count *= i
        print(count)

```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # Private variable
 
    def deposit(self, amount):
        self.__balance += amount
        print(f"₹{amount} deposited")
 
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"₹{amount} withdrawn")
        else:
            print("Insufficient Balance")
 
    def get_balance(self):
        return self.__balance
 
 
# Object Creation
acc = BankAccount(1000)
 
acc.deposit(500)
acc.withdraw(200)
 
print("Current Balance:", acc.get_balance())
```
