```python
class Student:
  def __init__(self,name,roll_no):
    self.name = name
    self.roll_no = roll_no
  # get
  def getName(self):
    print(self.name)
  # update
  def updateName(self):
    self.name = "hello"
    print(self.name)
  # delete
  def deleteName(self):
    del self.name
    # print(self.name)
  # create
  def createName(self,full_name):
    self.full_name = full_name
    print(self.full_name)
 
s = Student("nikhil",123)

s.createName()
```
