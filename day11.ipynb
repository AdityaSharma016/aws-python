```python
#filter
import numpy as np 
arr = np.array([10,20,30,40])
arr_f=arr[arr>20] #filter
print(arr_f)
```

```python
arr = np.array([10,20,6,30,40])
arr_f=arr[arr!=6]
print(arr_f)
```

```python
arr=np.array([1,2,3,4,5,12,15,16,20])
arr_f = arr[(arr >= 10) & (arr < 20)] 
print(arr_f)
```

```python
#fnacy indexing
arr=np.array([10,20,30,40,2,3,100])
print(arr)
arr_f = arr[[[0,2]]]
print(arr_f)
```

```python
#where 
arr=np.array([10,20,30,40,2,3,100])
print(arr)
arr_w=np.where(arr>9,"Trur","False")
print(arr_w)
```

```python

arr = np.array([10,20,3,4,14,15,16,90,100])
print(arr)
arr_w = np.where((arr>10) & (arr<20),arr*2,arr+2)
print(arr_w)

```

```python
#adding nd removing data 
arr1=np.array([10,20,30])
arr2=np.array([1,2,3])

 
arr1_arr2 =np.concatenate((arr1, arr2), axis=0)
print(arr1_arr2)

arr1=np.array([[10,20,30],[40,50,60]])
arr2=np.array([[1,2,3],[4,5,6]])
arr1_arr2=np.concatenate((arr1,arr2))
print(arr1_arr2)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
result=np.vstack((arr1,arr2))
print(result)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
result=np.hstack((arr1,arr2))
print(result)
```

```python
arr= np.array([2,3,4,5,6,7,8,45])
print(np.sum(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.max(arr))
print(np.min(arr))
```

```python
arr= np.array([2,3,4,5,6,7,8])
print(np.quantile(arr,0.3))
```

```python
#handling missing values
arr=np.array([10,20,np.nan,40])
print(arr)
#isnan is used to handle mising values
print(np.isnan(arr))
new_arr=np.nan_to_num(arr,nan=100)#null values ki jgh number print kra dega yh to awsome h yrr
print(new_arr)
```
