```python
#numpy array operations 
import numpy as np
arr= np.array([1,2,3,4,5])
print(arr[3])
```

```python
print(arr[0:4])#slicing 
print (arr[-1])
print(arr[:])
print(arr[0:4:2])#[start,stop,jump]
print(arr[::2])#jump by one
print(arr[::-1])#reverse can also be done by this 
print(arr[-2::])
```

```python
#2d arrays
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
print(arr[0][0])
print(arr[0][2])
print(arr[:,1]) #all row | cols 0
print(arr[0,0])
print(arr[:,2])

```

```python
arr=np.array([[10,20],[30,40]])
print(arr)
print(arr[0,0])
print(arr[1,1])
print(arr[:,1])
print(arr[[0,1],[1,0]])
print(arr[[0,1],[0,1]])
```

```python
#3d [block,row,column]
arr = np.array([[[2,3],[4,5]],[[8,9],[6,7]]])
print(type(arr))
arr.ndim
print(arr)
```

```python
#looping
arr = np.array([10,20,30,40,50])
print(arr)
for i in arr:
    print(i)
```
