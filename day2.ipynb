**operators**

2. Relational- >,<,<=,>=,==,!=
3. Assignment- +=,-=,*=,%=
4. logical- and and not  
5. bitwiese & ,|

```python
#arithmetic operations
a=10
b=20
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)# it gives the quotient without the decimal part
print(a%b)
print(a**b)# it gives the value of a raised to the power of b
```

```python
#Relational operators
print(a==b)
print(a!=b)
print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
```

```python
#logical operators
print(a>5 and b>15)
print(a>5 or b>15)
print(not(a>5 and b>15))
print(not(a>5 or b>15))
print(not(a>5) and not(b>15))
print(not(a>5) or not(b>15))
print(not(a>5) and b>15)
print(not(a>5) or b>15)
print(a>5 and not(b>15))
print(a>5 or not(b>15))
print(not(a>5) and not(b>15))
print(not(a>5) or not(b>15))
print(not(a>5) and not(b>15))
print(not(a>5) or not(b>15))    
```
