**conditional statement**- 
1. if
2. else
3. elif

```python
#if statement
num = 10 
if num > 2:
    print ("num=", num)
print (" if end")
```

```python
#if else
num1 =10
num2 = 20
if num1<=num2 : 
    print ("num1=", num1)
else :
    print ("num2=", num2)
```

```python
# if elif else
num1 = 10 
num2 = 20
num3 = 30
if num1 > num2 and num1 > num3 : 
    print ("num1=", num1)
elif num2 > num1 and num2 > num3 :
    print ("num2=", num2)
else :
    print ("num3=", num3)
    
```

```python
marks = int(input("Enter marks: "))
if marks >= 95 :
    print ("Grade A")
elif marks >= 90 :
    print ("Grade B")
elif marks >= 75 :
    print ("Grade C")
elif marks >= 60 :
    print ("Grade D")
else :
    print ("Fail")
```
